```python
# Mount Google Drive to access project files
from google.colab import drive
import os
```

```python
# 1. Mount Google Drive to the Colab environment
drive.mount('/content/drive')

# Define the target folder name and the search starting point
target_folder_name = 'Predicting Tool'
search_root = '/content/drive/MyDrive'

# 2. Recursively search for the directory
folder_path = None
for root, dirs, files in os.walk(search_root):
    # Check if the target folder exists within the current directory's subdirectories
    if target_folder_name in dirs:
        folder_path = os.path.join(root, target_folder_name)
        break # Exit the loop once the folder is found

# 3. Handle the result
if folder_path:
    print(f'Folder successfully found at: {folder_path}')

    # Change the current working directory to the located folder
    os.chdir(folder_path)
    print(f'Current working directory updated to: {os.getcwd()}')
    # List files in the current directory to verify model files
    print("Files in current directory:")
    for f in os.listdir(os.getcwd()):
        print(f)
else:
    print(f'Error: The folder "{target_folder_name}" was not found in Drive.')
```
